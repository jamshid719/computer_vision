```python
downloader.download("kebab food",
                   limit=100,
                   output_dir="/content/kebab_images",
                   adult_filter_off=True, # Bu adult filterni o‘chiradi.
                   force_replace=False, # eski rasmlarni o‘chirmaydi
                   timeout=60
                   )
```

```python
import os, shutil

source_dir = '/content/kebab_images/kebab food'
target_dir = '/content/drive/MyDrive/food101_dataset/kebab'

os.makedirs(target_dir, exist_ok=True)

for filename in os.listdir(source_dir): # har bir rasmlarni uqib target_dir ga saqlaydi
    src_file = os.path.join(source_dir, filename)
    dst_file = os.path.join(target_dir, filename)
    shutil.move(src_file, dst_file)
```
